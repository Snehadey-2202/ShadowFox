```python
# dice throw simulator(can be done using for loop and random library)

import random

dice = 0
prev_roll = 0
doublesix = 0
ans = [0] * 7

for i in range(100):
    dice = random.randint(1,6)
    ans[dice] += 1
    if(dice == 6 and prev_roll == 6):
        doublesix += 1
    prev_roll = dice

print("Total no. of rolls(1 to 6) are: ", ans[1:7])
print("Total no. of 1s: ",ans[1])
print("Total no. of 6s: ",ans[6])
print("Total no. of double 6s: ",doublesix)
```

```python
# jumping jacks count using for loop

for completed in range(10, 101, 10):
    print(f"\nPerform 10 jumping jacks.")

    if completed == 100:
        print("Congratulations! You completed the workout.")
        break

    tired = input("Are you tired? (yes/y or no/n): ").strip().lower()

    if tired == "yes" or tired == "y":
        skip = input("Do you want to skip the remaining sets? (yes/y or no/n): ").strip().lower()

        if skip == "yes" or skip == "y":
            print(f"You completed a total of {completed} jumping jacks.")
            break
        else:
            print(f"{100 - completed} jumping jacks are remaining.")
    else:
        print(f"{100 - completed} jumping jacks are remaining.")
```
