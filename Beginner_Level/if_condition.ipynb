```python
# BMI Category Calculator

weight = float(input("Enter your weight(in kg)"))
height = float(input("Enter your height(in meters)"))

BMI = weight/(height * height)

if(BMI >= 30):
    print("Obesity")
elif(BMI > 25 and BMI <= 29):
    print("Overweight")
elif(BMI > 18.5 and BMI <= 25):
    print("Normal")
else:
    print("Underweight")
```

```python
# if conditions to check if a city belongs to which country

Australia = ["Sydney", "Melbourne", "Brisbane", "Perth"]
UAE = ["Dubai", "Abu Dhabi", "Sharjah", "Ajman"]
India = ["Mumbai", "Bangalore", "Chennai", "Delhi"]

city = input("Enter a city name: ")
if(city in Australia):
    print("City is in Australia")
elif(city in UAE):
    print("City is in UAE")
elif(city in India):
    print("City is in India")
else:
    print("N/A")
```

```python
# if conditions to check if two cities belongs to same country

city1 = input("Enter a city name: ")
city2 = input("Enter another city name: ")

if(city1 in Australia and city2 in Australia):
    print("Both cities are in Australia")
elif(city1 in UAE and city2 in UAE):
    print("Both cities are in UAE")
if(city1 in India and city2 in India):
    print("Both cities are in India")
else:
    print("They do not belong to the same country")
```
