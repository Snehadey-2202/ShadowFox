```python
justice_league = ["Superman", "Batman", "Wonder Woman", "Flash", "Aquaman", "Green Lantern"]
```

```python
# len() method can be used to find out size of list, i.e., no. of members
print(len(justice_league))
```

```python
# append() method can be used to add members to the list

justice_league.append("Batgirl")
justice_league.append("Nightwing")

print(justice_league)
```

```python
# a combination of insert(), pop() and index() functions are used to move Wonder Woman to front of list

justice_league.insert(0, justice_league.pop(justice_league.index("Wonder Woman")))

print(justice_league)
```

```python
# similar to above operation

idx = justice_league.index("Flash")
justice_league.insert(idx + 1, justice_league.pop(justice_league.index("Green Lantern")))

print(justice_league)
```

```python
# we an use the slice method [:] to replace full list

justice_league[:] = ["Cyborg", "Shazam", "Hawkgirl", "Martian Manhunter", "Green Arrow"]

print(justice_league)
```

```python
# we can use inbuilt sort() function to sort alphabetically
justice_league.sort()

print(justice_league)
```

After sorting, hero at the 0th index is Cyborg. Thus Cyborg is the new leader.
